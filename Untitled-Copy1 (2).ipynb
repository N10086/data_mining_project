```python
from datetime import datetime

try:
    data = pd.read_csv('cleaned_data_modified.csv')
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please check if the file 'cleaned_data_modified.csv' exists and the format is correct.")
    exit()

# Check if required columns exist
required_columns = ['c_acc02e', 'c_acc03b', 'c_acc028']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    print(f"Error: The dataset is missing the following columns: {missing_columns}")
    exit()

# Define base date for conversion
base_date = datetime(2020, 1, 1)

# Function to convert date string to days since base_date
def convert_to_days(date_str):
    if pd.isna(date_str) or date_str == '\\N':
        return 0  # Replace missing values with 0
    try:
        date_obj = pd.to_datetime(date_str)
        days = (date_obj - base_date).days
        return int(days)
    except:
        return 0  # Handle invalid date formats

# Convert 'c_acc02e' and 'c_acc03b' to days since 2020-01-01
data['c_acc02e'] = data['c_acc02e'].apply(convert_to_days)
data['c_acc03b'] = data['c_acc03b'].apply(convert_to_days)

# Ensure the columns are of int type
data['c_acc02e'] = data['c_acc02e'].astype(int)
data['c_acc03b'] = data['c_acc03b'].astype(int)

# Save the modified data
data.to_csv('cleaned_data_modified.csv', index=False, encoding='utf-8-sig')
print("\nConverted 'c_acc02e' and 'c_acc03b' to the number of days since 2020-01-01 (integer type) and saved to 'cleaned_data_modified.csv'.")
```

### 2.6 Missing value handling
Missing values are handled differently in different models, and the details are discussed in the next section.

| Aspect                  | LSTM                              | Random Forest                     | SVM                               |
|-------------------------|-----------------------------------|-----------------------------------|-----------------------------------|
| **Imputation Method**   | Numeric features filled with 0   | Numeric features filled with 0   | Numeric features filled with median, categorical features filled with 'missing', date-derived features filled with mode |
| **Feature Type Handling** | No distinction, all features filled with 0 | No distinction, all features filled with 0 | Distinguishes numeric, categorical, and date features, with customized handling |
| **Target Missing Handling** | Not explicitly handled, assumes no missing values | Not explicitly handled, assumes no missing values | Drops rows with missing target values |
| **Date Feature Handling** | No special handling, filled with 0 | No special handling, filled with 0 | Split into year/month features, filled with mode |
| **Categorical Feature Handling** | No special handling, filled with 0 | No special handling, filled with 0 | Filled with 'missing' and encoded |
| **Complexity**          | Simple, uniform 0 fill           | Simple, uniform 0 fill           | Complex, tailored to feature types |
| **Model Characteristics Impact** | Requires complete input, 0 fill may introduce bias | Can natively handle missing values, but not utilized | Requires complete input, median fill more robust |
| **Applicable Scenarios** | Low missing rate, simple distribution | Low missing rate, accepts 0 fill | Moderate missing rate, mixed feature types |

Since columns with high missing values have already been deleted, filling missing values with 0 will not have a significant impact on model performance.

## 3.Model implementation

### 3.1 Model selection

Random Forest (RF):
- Robustly handles mixed features and high dimensionality with minimal preprocessing.
- Captures non-linear patterns and feature interactions for accurate predictions.
- Provides native Gini importance for ranking key predictors (e.g., edu_level).
- Mitigates class imbalance through ensemble averaging, ensuring balanced performance.
- Efficient and interpretable, ideal for static data and policy insights.


Support Vector Machine (SVM):
- Models non-linear patterns via kernels (e.g., RBF), suitable for complex relationships.
- Robust to outliers, enhancing reliability in noisy labor market data.
- Supports permutation importance for feature ranking, aiding policy recommendations.
- Excels in binary classification, aligning with the static dataset structure.

Long Short-Term Memory (LSTM):
- Captures potential temporal patterns in b_acc031, c_ajc090 via sequential modeling.
- Models complex, non-linear relationships, potentially uncovering subtle predictors.
- Supports SHAP-based feature importance for interpretable insights.
- Offers flexibility for future sequential data, enhancing academic innovation.

### Implemention of LSTM

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set font for Chinese display
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# Custom permutation importance for 3D LSTM input
def custom_permutation_importance(estimator, X, y, n_repeats=5, random_state=42, scoring='f1'):
    np.random.seed(random_state)
    n_features = X.shape[2]  # Number of features (3rd dimension)
    baseline_score = f1_score(y, estimator.predict(X))
    importances = np.zeros(n_features)
    
    for col in range(n_features):
        scores = []
        for _ in range(n_repeats):
            X_shuffled = X.copy()
            # Shuffle the specific feature across all samples
            X_shuffled[:, :, col] = np.random.permutation(X_shuffled[:, :, col])
            scores.append(f1_score(y, estimator.predict(X_shuffled)))
        importances[col] = baseline_score - np.mean(scores)
    
    return type('Result', (), {
        'importances_mean': importances,
        'importances_std': np.std([baseline_score - s for s in scores], axis=0)
    })()

# Load data
try:
    data = pd.read_csv('cleaned_data_modified.csv', encoding='utf-8-sig')
except FileNotFoundError:
    print("Error: cleaned_data_modified.csv file not found.")
    exit()
except Exception as e:
    print(f"Error loading data: {e}")
    exit()

# Drop the second row (index 1)
data = data.drop(index=1).reset_index(drop=True)

# Check number of columns
if data.shape[1] < 3:
    print("Error: Data must have at least 3 columns (ID column, at least 1 feature column, target column).")
    exit()

# Separate features and target
X = data.iloc[:, 1:-1]  # All columns except the first and last
y = data.iloc[:, -1]    # Last column as target
feature_names = X.columns.tolist()  # Store feature names for importance analysis

# Check if data is empty
if X.empty or y.empty:
    print("Error: Feature or target column is empty.")
    exit()

# Data preprocessing
# Convert to numeric, handle non-numeric values
X = X.apply(pd.to_numeric, errors='coerce')
if X.isna().any().any():
    print("Warning: Non-numeric values found in features, filled with 0.")
    X = X.fillna(0)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Prepare 3D input for LSTM (samples, timesteps, features)
X_reshaped = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

# Ensure target is binary (0/1)
y = y.astype(int)
if not set(y).issubset({0, 1}):
    print("Error: Target column contains non-0/1 values, only binary classification is supported.")
    exit()

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_reshaped, y, test_size=0.2, random_state=42, stratify=y
)

# Define LSTM model
def build_lstm_model(units1, units2, learning_rate):
    model = Sequential([
        LSTM(units1, input_shape=(1, X_scaled.shape[1]), return_sequences=True),
        Dropout(0.2),
        LSTM(units2),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Hyperparameter experiment configuration
base_params = {'units1': 50, 'units2': 25, 'learning_rate': 0.001, 'batch_size': 32}
experiments = [
    {
        'name': 'units1',
        'values': [30, 50, 70],
        'fixed': {'units2': 25, 'learning_rate': 0.001, 'batch_size': 32}
    },
    {
        'name': 'units2',
        'values': [15, 25, 35],
        'fixed': {'units1': 50, 'learning_rate': 0.001, 'batch_size': 32}
    },
    {
        'name': 'learning_rate',
        'values': [0.0001, 0.001, 0.01],
        'fixed': {'units1': 50, 'units2': 25, 'batch_size': 32}
    },
    {
        'name': 'batch_size',
        'values': [16, 32, 64],
        'fixed': {'units1': 50, 'units2': 25, 'learning_rate': 0.001}
    }
]

# Store results
results = []

# Run experiments
experiment_id = 1
for exp in experiments:
    param_name = exp['name']
    values = exp['values']
    fixed_params = exp['fixed']
    
    for value in values:
        # Construct current experiment parameters
        params = fixed_params.copy()
        params[param_name] = value

        
        # Build and train model
        model = build_lstm_model(
            params['units1'], params['units2'], params['learning_rate']
        )
        model.fit(
            X_train, y_train,
            epochs=10,
            batch_size=params['batch_size'],
            verbose=0,
            validation_split=0.2
        )
        
        # Predict
        y_pred_proba = model.predict(X_test, verbose=0)
        y_pred = (y_pred_proba > 0.5).astype(int).flatten()
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        
        # Store results
        results.append({
            'Experiment': experiment_id,
            'Parameters': params,
            'Tuned Parameter': f"{param_name}={value}",
            'Group': param_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        })      
        experiment_id += 1

# Summarize results
print("\n### Experiment Results Summary\n")
print("The following table summarizes the performance metrics for each experiment, with icons indicating performance levels: ✅ (>0.8), ⚠️ (0.6-0.8), ❌ (<0.6).\n")

# Create Markdown table
results_df = pd.DataFrame(results)

# Function to add icons based on metric value
def add_icon(value):
    if value > 0.8:
        return f"{value:.4f} "
    elif value >= 0.6:
        return f"{value:.4f} "
    else:
        return f"{value:.4f} "

# Define column widths for alignment
col_widths = {
    'Experiment': 10,
    'Group': 12,
    'Tuned Parameter': 18,
    'Accuracy': 12,
    'Precision': 12,
    'Recall': 12,
    'F1 Score': 12
}

# Build table header
table = f"| {'Experiment':<{col_widths['Experiment']}} | {'Group':<{col_widths['Group']}} | {'Tuned Parameter':<{col_widths['Tuned Parameter']}} | {'Accuracy':<{col_widths['Accuracy']}} | {'Precision':<{col_widths['Precision']}} | {'Recall':<{col_widths['Recall']}} | {'F1 Score':<{col_widths['F1 Score']}} |\n"
table += f"|{'-'*col_widths['Experiment']}-|{'-'*col_widths['Group']}-|{'-'*col_widths['Tuned Parameter']}-|{'-'*col_widths['Accuracy']}-|{'-'*col_widths['Precision']}-|{'-'*col_widths['Recall']}-|{'-'*col_widths['F1 Score']}-|\n"

# Add table rows
for _, row in results_df[['Experiment', 'Group', 'Tuned Parameter', 'Accuracy', 'Precision', 'Recall', 'F1 Score']].iterrows():
    table += f"| {row['Experiment']:<{col_widths['Experiment']}} | {row['Group']:<{col_widths['Group']}} | {row['Tuned Parameter']:<{col_widths['Tuned Parameter']}} | {add_icon(row['Accuracy']):<{col_widths['Accuracy']}} | {add_icon(row['Precision']):<{col_widths['Precision']}} | {add_icon(row['Recall']):<{col_widths['Recall']}} | {add_icon(row['F1 Score']):<{col_widths['F1 Score']}} |\n"

# Print table
print(table)

# Save results to CSV
results_df.to_csv('lstm_results.csv', index=False, encoding='utf-8-sig')
print("\nResults saved to lstm_results.csv")

# Visualize results (bar plots)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
for param_name in ['units1', 'units2', 'learning_rate', 'batch_size']:
    plt.figure(figsize=(12, 6))
    group_df = results_df[results_df['Group'] == param_name]
    x = np.arange(len(group_df))
    bar_width = 0.2
    
    for i, metric in enumerate(metrics):
        plt.bar(x + i * bar_width, group_df[metric], bar_width, label=metric)
    
    plt.xlabel(f'{param_name} Values')
    plt.ylabel('Score')
    plt.title(f'LSTM Experiment Metrics Comparison for Tuning {param_name}')
    plt.xticks(x + bar_width * 1.5, group_df['Tuned Parameter'])
    plt.legend()
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.savefig(f'lstm_{param_name}_metrics.png')
    plt.show()
    plt.close()
    print(f"\nMetrics comparison plot saved to lstm_{param_name}_metrics.png")

# Feature Importance Analysis
print("\n### Feature Importance Analysis\n")
# Select best model based on F1 Score
best_exp = results_df.loc[results_df['F1 Score'].idxmax()]
best_params = best_exp['Parameters']
print(f"Best Model (F1 Score {best_exp['F1 Score']:.4f}): {best_params}")

# Train best model
best_model = build_lstm_model(
    best_params['units1'], best_params['units2'], best_params['learning_rate']
)
best_model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=best_params['batch_size'],
    verbose=0,
    validation_split=0.2
)

# Custom wrapper for LSTM to work with permutation_importance
class LSTMWrapper:
    def __init__(self, model):
        self.model = model
    
    def fit(self, X, y, **kwargs):
        # Delegate to Keras model's fit method
        return self.model.fit(X, y, **kwargs)
    
    def predict(self, X):
        y_pred_proba = self.model.predict(X, verbose=0)
        return (y_pred_proba > 0.5).astype(int).flatten()

# Compute permutation importance
lstm_wrapper = LSTMWrapper(best_model)
perm_importance = custom_permutation_importance(
    lstm_wrapper, X_test, y_test, n_repeats=5, random_state=42, scoring='f1'
)

# Create DataFrame for importance scores
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Permutation Importance': perm_importance.importances_mean
})

# Normalize importance scores
importance_df['Permutation Importance'] /= importance_df['Permutation Importance'].sum()

# Sort by importance
importance_df = importance_df.sort_values('Permutation Importance', ascending=False)

# Print importance scores
print("\nFeature Importance Scores (Normalized):")
print(importance_df)

# Visualize feature importance
plt.figure(figsize=(10, 6))
x = np.arange(len(importance_df))
plt.bar(x, importance_df['Permutation Importance'], color='#1f77b4')
plt.xlabel('Features')
plt.ylabel('Normalised importance')
plt.title('LSTM Feature importance analysis')
plt.xticks(x, importance_df['Feature'], rotation=45, ha='right')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('lstm_feature_importance.png')
plt.show()
print("\n特征重要性图已保存为 lstm_feature_importance.png")
```

### Implemention of SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data
try:
    data = pd.read_csv('cleaned_data_modified.csv', encoding='utf-8-sig', na_values=['\\N'])
except FileNotFoundError:
    print("Error: cleaned_data_modified.csv file not found.")
    exit()
except Exception as e:
    print(f"Error loading data: {e}")
    exit()

# Drop the second row (index 1) if necessary
data = data.drop(index=1).reset_index(drop=True)

# Check number of columns
if data.shape[1] < 3:
    print("Error: Data must have at least 3 columns (ID, features, target).")
    exit()

# Identify target column (last column)
target_col = data.columns[-1]
print(f"Target column: {target_col}")

# Handle missing values in target
data[target_col] = data[target_col].replace(['\\N', ''], np.nan)
if data[target_col].isna().any():
    print(f"Warning: {data[target_col].isna().sum()} missing values in target, dropping rows.")
    data = data.dropna(subset=[target_col])

# Convert target to numeric
data[target_col] = pd.to_numeric(data[target_col], errors='coerce')
if data[target_col].isna().any():
    print("Error: Non-numeric values in target after conversion, dropping rows.")
    data = data.dropna(subset=[target_col])

# Ensure target is binary
data[target_col] = data[target_col].astype(int)
unique_classes = data[target_col].unique()
print(f"Target classes: {unique_classes}")
if len(unique_classes) < 2:
    print(f"Error: Target column contains only {len(unique_classes)} class(es): {unique_classes}")
    exit()
if not set(unique_classes).issubset({0, 1}):
    print("Warning: Target contains non-0/1 values, attempting to binarize.")
    data[target_col] = (data[target_col] > data[target_col].median()).astype(int)
    unique_classes = data[target_col].unique()
    print(f"New target classes after binarization: {unique_classes}")

# Separate features and target
X = data.drop(columns=['id', target_col])  # Exclude ID and target
y = data[target_col]
feature_names = X.columns.tolist()

# Check if data is empty
if X.empty or y.empty:
    print("Error: Feature or target data is empty.")
    exit()

# Preprocessing: Handle categorical and temporal features
for col in X.columns:
    if X[col].dtype == 'object':
        # Handle categorical columns
        try:
            if X[col].str.match(r'\d{4}-\d{2}-\d{2}').any():  # Detect date columns
                X[col] = pd.to_datetime(X[col], errors='coerce')
                X[f"{col}_year"] = X[col].dt.year.fillna(X[col].dt.year.mode()[0])
                X[f"{col}_month"] = X[col].dt.month.fillna(X[col].dt.month.mode()[0])
                X = X.drop(columns=[col])
            else:
                # Encode categorical variables
                X[col] = X[col].fillna('missing')
                le = LabelEncoder()
                X[col] = le.fit_transform(X[col])
        except AttributeError:
            # Handle non-string object columns
            X[col] = X[col].fillna('missing').astype(str)
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col])
    else:
        # Handle numeric columns
        X[col] = pd.to_numeric(X[col], errors='coerce').fillna(X[col].median())

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
feature_names = X.columns.tolist()  # Update feature names after preprocessing

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Verify class distribution in training set
train_classes = np.unique(y_train)
if len(train_classes) < 2:
    print(f"Error: Training set contains only {len(train_classes)} class(es): {train_classes}")
    exit()

# Define SVM model
def build_svm_model(C, kernel, gamma='scale'):
    model = SVC(
        C=C,
        kernel=kernel,
        gamma=gamma,
        random_state=42,
        probability=True,
        class_weight='balanced'  # Handle potential imbalance
    )
    return model

# Hyperparameter experiment configuration
base_params = {'C': 1.0, 'kernel': 'rbf', 'gamma': 'scale'}
experiments = [
    {
        'name': 'C',
        'values': [0.1, 1.0, 10.0],
        'fixed': {'kernel': 'rbf', 'gamma': 'scale'}
    },
    {
        'name': 'kernel',
        'values': ['linear', 'rbf', 'sigmoid'],
        'fixed': {'C': 1.0, 'gamma': 'scale'}
    },
    {
        'name': 'gamma',
        'values': ['scale', 'auto', 0.1],
        'fixed': {'C': 1.0, 'kernel': 'rbf'}
    }
]

# Store results
results = []

# Run experiments
experiment_id = 1
for exp in experiments:
    param_name = exp['name']
    values = exp['values']
    fixed_params = exp['fixed']
    
    for value in values:
        params = fixed_params.copy()
        params[param_name] = value
        print(f"\nExperiment {experiment_id}: {params}")
        
        # Build and train model
        model = build_svm_model(
            params['C'], params['kernel'], params.get('gamma', 'scale')
        )
        model.fit(X_train, y_train)
        
        # Predict
        y_pred = model.predict(X_test)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        
        # Store results
        results.append({
            'Experiment': experiment_id,
            'Parameters': params,
            'Tuned Parameter': f"{param_name}={value}",
            'Group': param_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        })
        
        experiment_id += 1

# Summarize results
print("\n### Experiment Results Summary\n")
print("The following table summarizes the performance metrics for each experiment, with icons indicating performance levels: ✅ (>0.8), ⚠️ (0.6-0.8), ❌ (<0.6).\n")

# Create Markdown table
results_df = pd.DataFrame(results)

# Function to add icons based on metric value
def add_icon(value):
    if value > 0.8:
        return f"{value:.4f} "
    elif value >= 0.6:
        return f"{value:.4f} "
    else:
        return f"{value:.4f} "

# Define column widths for alignment
col_widths = {
    'Experiment': 10,
    'Group': 12,
    'Tuned Parameter': 18,
    'Accuracy': 12,
    'Precision': 12,
    'Recall': 12,
    'F1 Score': 12
}

# Build table header
table = f"| {'Experiment':<{col_widths['Experiment']}} | {'Group':<{col_widths['Group']}} | {'Tuned Parameter':<{col_widths['Tuned Parameter']}} | {'Accuracy':<{col_widths['Accuracy']}} | {'Precision':<{col_widths['Precision']}} | {'Recall':<{col_widths['Recall']}} | {'F1 Score':<{col_widths['F1 Score']}} |\n"
table += f"|{'-'*col_widths['Experiment']}-|{'-'*col_widths['Group']}-|{'-'*col_widths['Tuned Parameter']}-|{'-'*col_widths['Accuracy']}-|{'-'*col_widths['Precision']}-|{'-'*col_widths['Recall']}-|{'-'*col_widths['F1 Score']}-|\n"

# Add table rows
for _, row in results_df[['Experiment', 'Group', 'Tuned Parameter', 'Accuracy', 'Precision', 'Recall', 'F1 Score']].iterrows():
    table += f"| {row['Experiment']:<{col_widths['Experiment']}} | {row['Group']:<{col_widths['Group']}} | {row['Tuned Parameter']:<{col_widths['Tuned Parameter']}} | {add_icon(row['Accuracy']):<{col_widths['Accuracy']}} | {add_icon(row['Precision']):<{col_widths['Precision']}} | {add_icon(row['Recall']):<{col_widths['Recall']}} | {add_icon(row['F1 Score']):<{col_widths['F1 Score']}} |\n"

# Print table
print(table)

# Save results to CSV
results_df.to_csv('svm_results.csv', index=False, encoding='utf-8-sig')
print("\nResults saved to svm_results.csv")

# Visualize results (bar plots)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
for param_name in ['C', 'kernel', 'gamma']:
    plt.figure(figsize=(12, 6))
    group_df = results_df[results_df['Group'] == param_name]
    x = np.arange(len(group_df))
    bar_width = 0.2
    
    for i, metric in enumerate(metrics):
        plt.bar(x + i * bar_width, group_df[metric], bar_width, label=metric)
    
    plt.xlabel(f'{param_name} Values')
    plt.ylabel('Score')
    plt.title(f'SVM Experiment Metrics Comparison for Tuning {param_name}')
    plt.xticks(x + bar_width * 1.5, group_df['Tuned Parameter'])
    plt.legend()
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.savefig(f'svm_{param_name}_metrics.png')
    plt.show()
    plt.close()
    print(f"\nMetrics comparison plot saved to svm_{param_name}_metrics.png")

# Feature Importance Analysis
print("\n### Feature Importance Analysis\n")
# Select best model based on F1 Score
best_exp = results_df.loc[results_df['F1 Score'].idxmax()]
best_params = best_exp['Parameters']
print(f"Best Model (F1 Score {best_exp['F1 Score']:.4f}): {best_params}")

# Train best model
best_model = build_svm_model(
    best_params['C'], best_params['kernel'], best_params.get('gamma', 'scale')
)
best_model.fit(X_train, y_train)

# Compute permutation importance
perm_importance = permutation_importance(
    best_model, X_test, y_test, n_repeats=10, random_state=42, scoring='f1'
)

# Compute SVM coefficients if linear kernel
svm_coeff = None
if best_params['kernel'] == 'linear':
    svm_coeff = np.abs(best_model.coef_[0])  # Absolute coefficients for binary classification
    svm_coeff /= svm_coeff.sum()  # Normalize

# Create DataFrame for importance scores
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Permutation Importance': perm_importance.importances_mean
})
if svm_coeff is not None:
    importance_df['SVM Coefficient'] = svm_coeff

# Normalize permutation importance
importance_df['Permutation Importance'] /= importance_df['Permutation Importance'].sum()

# Sort by Permutation Importance
importance_df = importance_df.sort_values('Permutation Importance', ascending=False)

# Print importance scores
print("\nFeature Importance Scores (Normalized):")
print(importance_df)

# Visualize feature importance
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = np.arange(len(importance_df))
plt.bar(x - bar_width/2, importance_df['Permutation Importance'], bar_width, label='Permutation Importance', color='#1f77b4')
if svm_coeff is not None:
    plt.bar(x + bar_width/2, importance_df['SVM Coefficient'], bar_width, label='SVM Coefficient', color='#ff7f0e')
plt.xlabel('Features')
plt.ylabel('Normalised importance')
plt.title('SVM Feature importance analysis')
plt.xticks(x, importance_df['Feature'], rotation=45, ha='right')
plt.legend()
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('svm_feature_importance.png')
plt.show()
print("\n特征重要性图已保存为 svm_feature_importance.png")
```
